```python
import pandas as pd
import PyPDF2
import re
import os
from sentence_transformers import SentenceTransformer, util

# **Load Sentence-BERT Model for Semantic Scoring**
model = SentenceTransformer("all-MiniLM-L6-v2")

# **Function to extract answers from a PDF file**
def extract_answers_from_pdf(pdf_path):
    with open(pdf_path, "rb") as file:
        reader = PyPDF2.PdfReader(file)
        text = ""

        for page in reader.pages:
            page_text = page.extract_text()
            if page_text:
                text += page_text + " "  # Concatenate text from all pages

    # **Fix: Split answers based on numbering (e.g., 1., 2., 3.)**
    answers = re.split(r'\d+\.\s*', text)  # Splits at "1. ", "2. ", etc.
    answers = [ans.strip() for ans in answers if ans.strip()]  # Clean up extra spaces

    return answers

# **Function to calculate similarity score**
def calculate_score(student_answer, expected_answer):
    if pd.isna(student_answer) or pd.isna(expected_answer):
        return 0  # If missing, assign zero score
    student_embedding = model.encode(student_answer, convert_to_tensor=True)
    expected_embedding = model.encode(expected_answer, convert_to_tensor=True)
    similarity = util.pytorch_cos_sim(student_embedding, expected_embedding).item()
    return round(similarity * 100, 2)  # Convert to percentage

# **Load the teacher's dataset (Expected Answers)**
dataset_path = r"D:\Users\abc\Downloads\fresh.csv"  # Modify if needed
df = pd.read_csv(dataset_path)

# Ensure dataset has required columns
if "Question" not in df.columns or "Expected Answer" not in df.columns:
    raise ValueError("Dataset is missing required columns: 'Question' or 'Expected Answer'")

# **Folder containing student PDFs**
student_pdfs_folder = r"D:\Users\abc\Downloads\All students" # Update this path as needed

# **Process each student's PDF in batch**
scores = {}  # Dictionary to store scores for each student

for filename in os.listdir(student_pdfs_folder):
    if filename.endswith(".pdf"):
        pdf_path = os.path.join(student_pdfs_folder, filename)
        student_answers = extract_answers_from_pdf(pdf_path)

        # Ensure answer count matches question count
        if len(student_answers) == len(df["Question"]):
            student_name = filename.replace(".pdf", "")  # Extract student name from filename
            df[student_name] = student_answers  # Append student answers

            # **Calculate scores for each answer**
            score_column = f"{student_name}_Score"
            df[score_column] = df.apply(lambda row: calculate_score(row[student_name], row["Expected Answer"]), axis=1)
            scores[student_name] = df[score_column].mean()  # Calculate average score per student

        else:
            print(f"⚠️ Warning: Mismatch in number of answers for {filename}. Check extracted answers.")

# **Save the graded dataset**
graded_dataset_path = r"D:\Users\abc\Downloads\graded_question_answer_pairs.csv"
df.to_csv(graded_dataset_path, index=False)

print(f"✅ Batch processing & grading completed. Graded dataset saved at: {graded_dataset_path}")

# **Save summary scores**
summary_scores_path = r"D:\Users\abc\Downloads\student_scores_summary.csv"
pd.DataFrame(scores.items(), columns=["Student", "Average Score"]).to_csv(summary_scores_path, index=False)

print(f"✅ Student scores summary saved at: {summary_scores_path}")
```
